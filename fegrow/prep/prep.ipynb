```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
molecules = []
with Chem.SDMolSupplier('ligands.sdf') as SDF:
    # for each mol
    for mol in SDF:
        if mol is None:
            continue
        molecules.append(mol)

        
len(molecules)
```

```python
for mol in molecules:
    smi = Chem.MolToSmiles(mol)
    print(smi)
```

```python
sdfFile = open('core.sdf', 'a')
w = Chem.SDWriter(sdfFile)
w.write(molecules[15])
w.close()
```
